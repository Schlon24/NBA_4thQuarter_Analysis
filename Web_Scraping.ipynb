## Web scraping from stats.nba.com

```python
from pandas.io.html import read_html
import numpy as np
import pandas as pd
from selenium import webdriver
```

```python
#Set the webdriver

driver=webdriver.Chrome("C:/ProgramData/chromedriver.exe")
```

```python
#The target website, where you can find all kind of stats about the league
#First, I want the advanced box score stats from all games in the 2019/2020 season

driver.get("https://stats.nba.com/teams/boxscores-advanced/?Season=2019-20&SeasonType=Regular%20Season")
```

```python
#The website has 39 pages, 50 rows per page
#There is a dropdown to select the pages, but it has also the option "All" to show all rows in one page
#So in the elements I copied the xpath for the dropdown selection "All" 

all_pages=driver.find_element_by_xpath("/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[1]/div/div/select/option[1]")
```

```python
#Now we can start to scrape
#After execute the code: the webdriver is selecting all pages in the dropdown
#then the webdriver finds the table,
#convert the table from JavaScript to Html.
#now pandas can convert it to a Dataframe

all_pages.click()
table=driver.find_elements_by_class_name("nba-stat-table") #from the elements, the class name of the table
table_html = table[0].get_attribute('innerHTML')
df = read_html(table_html)[0]
    
driver.quit()    
```

```python
#control of the data:
#The data looks good, organized and clean
#only little data cleaning necessary

season_2019=df
season_2019.head()
```

```python
#the data is also complete, scraped all rows and columns

season_2019.shape
```

```python
#The same steps are for the 4th Quarter stats
#Only now I selected the 4th Quarter filter at the website manually and copied the new Url

driver.get("https://stats.nba.com/teams/boxscores-advanced/?Season=2019-20&SeasonType=Regular%20Season&Period=4")
all_pages_4Q=driver.find_element_by_xpath("/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[1]/div/div/select/option[1]")
all_pages_4Q.click()
table=driver.find_elements_by_class_name("nba-stat-table")
table_html = table[0].get_attribute('innerHTML')
df = read_html(table_html)[0]
    
driver.quit()  
```

```python
season_2019_4Q=df
season_2019_4Q.head()
```

```python
#Saved the dataframes to a csv file

season_2019.to_csv("season_2019.csv",index = False)
season_2019_4Q.to_csv("season_2019_4Q.csv", index = False)
```
